```python
import pandas as pd
import numpy as np
```

```python
import random
```

```python
L = []
for i in range(10000):
  a = random.randint(1,6)
  b = random.randint(1,6)

  L.append(a + b)
```

```python
len(L)
```

```python
L[:5]
```

```python
s = (pd.Series(L).value_counts()/pd.Series(L).value_counts().sum()).sort_index()
```

```python
s.plot(kind='bar')
```

```python
np.cumsum(s).plot(kind='bar')
```
