### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatterplot -> axes level plot
sns.scatterplot(data= tips, x='total_bill', y='tip')
```

```python
# figure level function 
# scatterplot -> axes level plot
sns.relplot(data= tips, x='total_bill', y='tip', style='sex', hue='time',size='sex')
```

```python
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
gap = px.data.gapminder()
gap
```

```python
data = gap[gap['country'] == 'India']
```

```python
# axes level function 
sns.lineplot(data = data, x = 'year' , y= 'lifeExp')
```

```python
sns.relplot(data = data, x='year', y='lifeExp', kind='line')
```

```python
data = gap[gap['country'].isin(["India", "China", "Pakistan"])]
```

```python
sns.relplot(data=data, x = 'year', y='lifeExp', hue='country', kind='line')
```

### facet plot

```python
tips
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex' , row='smoker')
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level functio -> displot
# axes level function -> histplot, kdeplot, rugplot
```

```python
sns.histplot(data = tips, x = 'total_bill')
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist')
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins=20) 
```

```python
sns.displot(data=tips, x = 'day', kind='hist')
```

```python
sns.displot(data=tips, x = 'tip', kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step',hue='sex')
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step',col='sex')
```

```python
sns.kdeplot(data=titanic, x='age',hue='sex')
```

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age')
```

```python
sns.kdeplot(data=tips, x = 'total_bill',y='tip')
```

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
