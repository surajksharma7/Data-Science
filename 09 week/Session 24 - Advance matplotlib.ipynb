```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Iris

```python
iris = pd.read_csv("iris.csv")
iris
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
```

```python
iris['value'] = iris['Species']
```

```python
iris['value'].unique()
```

```python
iris['value'].replace({'Iris-setosa':1,'Iris-versicolor':2, 'Iris-virginica':3}, inplace=True)
```

```python
plt.figure(figsize=(7,3))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c= iris['value'],cmap='summer',alpha=0.9)
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
plt.show()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
```

```python
batters.shape
```

```python
df = batters.head(100).sample(25, random_state=5)
```

```python
df.sample()
```

```python
plt.scatter(df['avg'], df['strike_rate'], s= df['runs']*.25)
for i in range(df.shape[0]):
    plt.text(df['avg'].iloc[i],df['strike_rate'].iloc[i],df['batter'].iloc[i])
plt.axhline(df['strike_rate'].mean())
plt.axvline(df['avg'].mean())
plt.show()
```

### Subplots

```python
fig, ax = plt.subplots()
ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, sharex = True)


ax[0].scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax[1].scatter(batters['avg'], batters['runs'],marker='+')

ax[0].set_title('avg vs strike rate')
ax[1].set_title('avg vs runs')
ax[1].set_xlabel('avg')
```

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100,100])
ax.plot3D(x,y,z,color='red')
plt.show(
)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
z.shape

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.colorbar()
plt.plot()
```
