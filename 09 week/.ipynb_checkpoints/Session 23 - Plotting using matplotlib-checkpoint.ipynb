```python
# from a pandas dataframe

batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman["index"],batsman['RG Sharma'],batsman["index"],batsman['V Kohli'])
```

```python
# plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'], label="rohit sharma")
plt.plot(batsman['index'],batsman['RG Sharma'], label = "virat kohli")
plt.legend()
```

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# grid
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()

plt.show()
```

```python
# limiting axes
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color=colors)
```

```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','blue','green','yellow','pink']

plt.barh(colors,children,color=colors)
```

```python
df = pd.read_csv("batsman_season_record.csv")
df
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2, label = '2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2, label = '2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2, label = '2017')
plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.legend()
plt.show()
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2, label = '2015')
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2, label = '2016')
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2, label = '2017')
plt.xticks(np.arange(df.shape[0]), df['batsman'], rotation='vertical')
plt.legend()
plt.show()
```

```python
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'],bottom = df['2015'])
plt.bar(df['batsman'], df['2017'], bottom = (df['2015'] + df['2016']))
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23,45,100,20,49]
subjects = ['eng','science','maths','sst','hindi']
plt.pie(data,labels=subjects)

plt.show()
```

```python
df = pd.read_csv("gayle-175.csv")
df
```

```python
plt.pie(df['batsman_runs'], labels= df['batsman'], autopct = '%0.1f%%',explode=[0.1,0,0,0,0,0])
plt.show()
```
