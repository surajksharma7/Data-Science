### pivot Table

The Poivot table is a simple column-wise data as input, and groups the entries into a two dimensional table that provides a multidimensional summarization of the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

df = sns.load_dataset('tips')
df

```python
df.groupby('sex')[['total_bill']].mean()
```

```python
df.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```

```python
df.pivot_table(index='sex', columns = 'smoker',values = 'total_bill', aggfunc='count')
```

```python
# what if we dont specify values
df.pivot_table(index='sex', columns = 'smoker',aggfunc='count')
```

```python
# multidimensional 
df.pivot_table(index = ['sex','smoker'], columns=['day','time'], values = 'total_bill',aggfunc='mean')
```

```python
# multidimensional 
df.pivot_table(index = ['sex','smoker'], columns=['day','time'], aggfunc={'size':'mean','tip':'max','total_bill':'mean'})
```

```python
# multidimensional 
df.pivot_table(index = ['sex','smoker'], columns=['day','time'], aggfunc='sum', margins=True)
```
